### Import packages

```python
import requests
from bs4 import BeautifulSoup
import json
```

### Load the page

```python
page = requests.get("https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number")
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.contents)
```

### Find tables containing Pokemon details

```python
poke_content = soup.find(id = 'mw-content-text')
poke_tables = poke_content.find_all('table')
```

Update poke_tables so that it only includes tables containing information on Pokemon

```python
poke_tables = poke_tables[1:9]
```

Look for indices containing relevant information

```python
poke_tables[0].contents[3]
```

```python
info_start = 3
info_row = poke_tables[0].contents[info_start]
```

```python
for i in range(len(info_row.contents)):
    print(f'{i} - {info_row.contents[i]}'.replace('\n', ''))
```

From what we can see, indices 1(kdex), 3(ndex), 7(name and URL), 9(type1), and 11(type2) contain the data that we need to extract.

```python
kdex=info_row.contents[1].text.strip()
ndex=info_row.contents[3].text.strip()
name=info_row.contents[7].text.strip()
type1=info_row.contents[9].text.strip()
type2=info_row.contents[11].text.strip()
pURL = 'https://bulbapedia.bulbagarden.net' + info_row.contents[7].a.get('href')

print(f'Pokemon {ndex} {name} is a {type1} and {type2} Pokemon. {pURL}.')
```

```python
for j in range(len(poke_tables)):
    print(f'Generation {j+1}')
    for i in range(info_start, len(poke_tables[j].contents), 2):
        poke_info=poke_tables[j].contents[i]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        type1=poke_info.contents[9].text.strip()
        pURL = 'https://bulbapedia.bulbagarden.net' + info_row.contents[7].a.get('href')
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            print(f'Pokemon {ndex} {name} is a {type1} & {type2} Pokemon')
        else:
            print(f'Pokemon {ndex} {name} is a {type1} Pokemon')
        print(f'URL = {pURL}')
```

### Parse to JSON object

```python
generations = []

for j in range(len(poke_tables)):
    data = []
    for i in range(info_start, len(poke_tables[j].contents), 2):
        poke_info=poke_tables[j].contents[i]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        type1=poke_info.contents[9].text.strip()
        pURL = 'https://bulbapedia.bulbagarden.net' + info_row.contents[7].a.get('href')
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            data.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2,
                "pURL" : pURL
            })
        else:
            data.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "pURL" : pURL
            })
    generations.append({
        "generation": j+1,
        "data": data
    })
    
generations
```

### Save to JSON file

```python
with open('pokemon.json', 'w') as json_file:
    json.dump(generations, json_file)
```
